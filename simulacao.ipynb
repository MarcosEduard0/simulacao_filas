```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# num_samples = int(input("Digite o número de amostras: "))
# mean_arrival_time = float(input("Digite o tempo médio de chegada das amostras: "))
# mean_departure_time = float(input("Digite o tempo médio de saída das amostras: "))

num_samples = 120
mean_arrival_time = 1
mean_departure_time = 2
```

```python
print("\nTempo de chegada | Tipo de evento | Número de amostras\n")

n = 0
simul_time = 0
data = []

# To calculate waiting line mean time
arrivals = []
waiting_line_time = 0

for i in range(num_samples):
    arrival_time = np.random.exponential(scale=1/mean_arrival_time)
    departure_time = np.random.exponential(scale=1/mean_departure_time)

    if n == 0 or arrival_time < departure_time:
        simul_time += arrival_time
        print(f"{simul_time:8.4f} \t\t{'Chegada':14} {n:2} => {n+1:2}")
        data.append([f"{simul_time:8.4f}", f"{'Chegada'}", f"{n}", f"{n+1}"])
        n += 1
        if (n > 1):
            arrivals.append(simul_time)
    else:
        simul_time += departure_time
        if (len(arrivals) > 0):
            waiting_line_time += simul_time - arrivals[0]
            arrivals.pop(0)
        print(f"{simul_time:8.4f} \t\t{'Saída':14} {n:2} => {n-1:2}")
        data.append([f"{simul_time:8.4f}", f"{'Saída'}", f"{n}", f"{n+1}"])
        n -= 1

print(f"\nTotal de pessoas que foram ao banco na simulação: {num_samples}")
print(f"Número de pessoas na fila ao final da simulação: {n}")
print(f"Tempo total da simulação: {simul_time:.4f}")
print(f"Tempo médio de espera na fila: {(waiting_line_time/num_samples):.4f}")
```

```python
df = pd.DataFrame(data, columns=['Tempo de Chegada', 'Tipo de Evento', 'Antes', 'Depois'])
df
```
